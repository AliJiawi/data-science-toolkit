```python
import numpy as np
```

```python
array_2x2 = np.array([[1, 2], [3, 4]])
print(array_2x2)
```

```python
array_random_3x4 = np.random.rand(3, 4)
print(array_random_3x4)
```

```python
array_ones = np.ones((2, 4))
array_zeros = np.zeros((2, 4))
print(array_ones)
print(array_zeros)
```

```python
array_random_6x3 = np.random.rand(6, 3)
print("Random 6x3 array:")
print(array_random_6x3)
array_random_6x3_transposed = array_random_6x3.T
print("\nTransposed random 6x3 array:")
print(array_random_6x3_transposed)
```

```python
n = 2
p = 4
m = 8

array_nxp = np.random.rand(n, p)
array_pxm = np.random.rand(p, m)
array_dot_product = np.dot(array_nxp, array_pxm)

print("Array n x p")
print(array_nxp)
print("\nArray p x m")
print(array_pxm)
print("\nDot Product")
print(array_dot_product)
```
